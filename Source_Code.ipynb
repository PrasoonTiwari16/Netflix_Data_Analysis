### Show only the title of all shows that were relesded in india only

```python
df.columns
```

```python
# Assuming 'india_tv_shows' contains the filtered DataFrame
india_tv_shows = df[(df['Category'] == 'TV Show') & (df['Country'] == 'India')]['Title']

# Calculate the padding needed
padding = 3 - (len(india_tv_shows) % 3)

# Pad the array with NaN values
padded_tv_shows = np.append(india_tv_shows.values, [np.nan] * padding)

# Reshape the padded data into a 3x3 matrix
matrix_data = padded_tv_shows.reshape((-1, 3))

# Create a new DataFrame with the reshaped data
matrix_df = pd.DataFrame(matrix_data, columns=['Column 1', 'Column 2', 'Column 3'])

# Display the matrix DataFrame
print(matrix_df)
```

### In how many movies/shows, Tom Cruise was cast.

```python
# Filter DataFrame by category 'Movie' and 'TV Show'
filtered_df = df[df['Category'].isin(['Movie', 'TV Show'])]

# Check if 'Tom Cruise' is in the 'Cast' column
tom_cruise_movies = filtered_df[filtered_df['Cast'].str.contains('Tom Cruise', na=False)]

# Count the number of movies/shows
count = len(tom_cruise_movies)

# Display the count
print("Number of movies/shows with Tom Cruise:", count)
```

```python
tom_cruise_movies
```

```python
# Get unique directors of Tom Cruise movies/shows
directors = tom_cruise_movies['Director'].unique()

# Display the directors
print("Directors of movies/shows with Tom Cruise:")
for director in directors:
    print(director)
```
